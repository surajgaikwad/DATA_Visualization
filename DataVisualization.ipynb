```python
"""
use NumPy to generate random data 
"""
import numpy as np

x = np.linspace(0, 10, 25)
y = x*x+2
print(x)
print(y)
print(np.array([x,y]).reshape(25,2))
```

```python
"""
Drawing a subgraph
"""
pylab.subplot(1,2,1) #contents of the bracket represent row , column , index
pylab.plot(x,y, 'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x, 'g*-')
```

```python
# (1,2,1) and (1,2,2) is necesaary to get the plots in different plates or the subplots will fall on each other 
# for example 
pylab.subplot(1,2,1) #contents of the bracket represent row , column , index
pylab.plot(x,y, 'r--')
pylab.subplot(1,2,1)
pylab.plot(y,x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axis=fig.add_axes([0.5 , 0.1, 0.8, 0.8]) #controls the left , bottom , width , height of the canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
# again we can draw subplots 
fig, axes = plt.subplots(nrows=1, ncols=2) # submap is of 1 row and 2 columns 
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# we can also draw a graph inside another graph 
fig = plt.figure()
#controls the left , bottom , width , height of the canvas (from 0 to 1)
axes1=fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes
axes2 = fig.add_axes([0.1, 0.6, 0.4, 0.3]) #small canvas # change these numbres to change the location of small graph on big graph

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
n = np.array([0,1,2,3,4,5])

fig, axes= plt.subplots(1, 4, figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x, x+0.25*np.random.random(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2, lw=2)

axes[2].set_title("bar")
axes[2].bar(n, n**2, align="center", alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

```python
"""
Draw a Raddar chart
"""
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, 0.6, 0.6], polar=True)
t = np.linspace(0, 2*np.pi, 100)
ax.plot(t,t, color='blue', lw=3) # just a diagonal line in a regular chart
ax.plot(t,t*0.5, color='red', lw=3) 
```

```python
"""
Draw a Histogram
"""
n = np.random.randn(10000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("cumulative Detailed Histogram")
axes[1].hist(n, cumulative=True, bins=10, alpha=0.8)
```

```python
# Draw Contour Image

import matplotlib
import numpy as np
import matplotlib.cm as cm        #cm for colormap
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1-Z2) * 2

print(X)
print(Y)
```

```python
fig , ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour map')
```

```python
# Draw a 3D Surface

from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))

# specify the 3D graphics to draw , with projection ='3d'
ax = fig.add_subplot(1,2,1, projection= '3d')
ax.plot_surface(X,Y,Z, rstride=4, cstride=4, linewidth=0)
```

# Write a Python Program to create a pie chart of popularity of programming languages

```python
import matplotlib.pyplot as plt
# Data to plot
languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
```

```python
# explode 1st slice
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(popularity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90 )
```
