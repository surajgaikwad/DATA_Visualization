#  Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mtcars = pd.read_csv('C:/Users/Expert/Desktop/py4e/DataVisualization/mtcars.csv')
```

```python
mtcars.head() #mtcars data is taken from 1974 motor trend US magzine , contains info about fuel consumption and 11 different aspects of 32 cars
# model, mpg=miles per gallon , cyl=number of cylinders , disp=displacement in inch^3, hp=horse power, drat=rear axel tation , wt = wt in pounds ....etc
```

```python
mtcars.info()
```

```python
mtcars.shape
```

```python
#barplot
res=sns.barplot(mtcars['cyl'], mtcars['mpg'])
plt.show()
```

```python
res = sns.barplot(x='cyl' , y='mpg' , data=mtcars)
plt.show
```

```python
res = sns.barplot(x='cyl', y='mpg', data=mtcars, color='yellow')
plt.show()
```

```python
#assigning palettes fo different colors
res = sns.barplot(x='cyl', y='mpg', data=mtcars, color='yellow', palette="Set2")
plt.show()
```

```python
res = sns.barplot(x='gear', y='hp', data=mtcars, color='yellow', palette="rocket")
plt.show()
```

```python
# Countplot

sns.countplot(x = 'cyl', data = mtcars, palette="Set1") #make sure the data name is without qoutes
```

```python
sns.countplot(x = 'carb', data = mtcars, palette="Set2") #make sure the data name is without qoutes
```

```python
# horizontal countplots 
sns.countplot(y = 'gear', data=mtcars, palette="Set2")
```

```python
# hue parameter
sns.countplot(x = 'gear', hue = 'cyl', data=mtcars, palette="Set1")
```

```python
#distplot =distribution plot
sns.distplot(mtcars.mpg, bins = 10, color = 'g')
```

```python
# heatmap using seabonr , makes you view matrix like data , correlation maps 

sns.heatmap(mtcars.corr(), cbar = True, linewidth=0.5)

#lighter color positive correlation , dark negative correlation , 
# negative correlation example = number of cylinder increases , mpg decreases
```
