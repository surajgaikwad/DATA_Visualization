```python
opsd_daily = pd.read_csv('opsd_germany_daily.csv', index_col=0, parse_dates=True) # all above steps in one line code
opsd_daily
```

```python
# Add columns with year, month, and weekday name
opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month                     # for month name - opsd_daily.index.month_name()
opsd_daily['Weekday Name'] = opsd_daily.index.day_name()         # notice the () just for day_name method 
# Display a random sampling of 5 rows
opsd_daily.tail(5)
```

### Customizing time series plots

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots()
ax.plot(opsd_daily.loc['2017-01':'2017-02', 'Consumption'], marker='o', linestyle='--')
ax.set_ylabel("Daily Consumption GWh")
ax.set_title("Jan-Feb Electricity COnsumption")

# Ste x-axis major ticks to weekly interval, on monday (vertial white line on Monday)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))

# Format x-ticks labels as 3 letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
```
